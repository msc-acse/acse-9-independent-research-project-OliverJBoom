## Loading Data

```python
path = "Data/"
universe_dict = universe_select(path, "Cu")
```

## Preprocessing

```python
# Renaming the columns to price
universe_dict = price_rename(universe_dict)
# Cleaning the dataset of any erroneous datapoints
universe_dict = clean_dict_gen(universe_dict)
# Making sure that all the points in the window have consistent lenght
universe_dict = truncate_window_length(universe_dict)
# Generate the full training dataset
df_full = generate_dataset(universe_dict, lag=5, lg_returns_only=False)
```

The target column represents the log returns at one forecast length out in the future. This is as in the auto-regressive calculations log returns will be used as the sole input variable, so naming convention has been kept separate to avoid confusion.

So if the forecast length is a one week prediction, then the log returns for +5 days will be the target variable. 

```python
# Visualise the plots
# visualise_universe(universe_dict)
df = df_full[["target_cu_lme"]]
```

```python
df_full.head(5)
# df.head()
```

```python
# Taking t-1 to be the value for t
df["persistance"] = df.shift(1)

# Calculating metrics for these columns
MSE, MAE, MDE = evaluate(df, "target_cu_lme", "persistance")
```

```python
# Placing in results dataframe
results = pd.DataFrame(columns={"MSE", "MAE", "MDE"})
results.index.name = 'Name'
results.head()
results.loc["persistance"] = [MSE, MAE, MDE] 
results.head()
```
